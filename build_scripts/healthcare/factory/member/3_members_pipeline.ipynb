```python
sample_size = 10000
seed = None
```

-------------------------
## Synthetic Builder

```python
builder = SyntheticBuilder.from_env('members')
```

### Run the Pipeline  save the Schema and show the Canonical Report

```python
df = builder.load_persist_canonical()
```

```python
# set the version
builder.set_version(VERSION)

# run the pipeline
builder.run_component_pipeline(size=sample_size)

#save schema
builder.save_canonical_schema()

# report
builder.canonical_report(builder.load_persist_canonical(), stylise=True)
```

### Save Reports

```python
reports = [Commons.param2dict(report=builder.REPORT_CATALOG, file_format='csv', versioned=True)]
builder.save_report_canonical(reports=reports, report_canonical=builder.report_column_catalog(stylise=False))
```
