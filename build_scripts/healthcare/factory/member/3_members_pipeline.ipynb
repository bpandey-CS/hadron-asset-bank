```python
sample_size = 10000
seed = None
```

-------------------------
## Synthetic Builder

```python
builder = SyntheticBuilder.from_env('members')
```

### Run the Pipeline  save the Schema and show the Canonical Report

```python
# set the version
builder.set_version(VERSION)

# run the pipeline
builder.run_synthetic_pipeline(size=sample_size)

# save the schama from the pipeline outcome
builder.save_canonical_schema()

# report the stored schema
builder.report_canonical_schema(stylise=True)
```

### Save Reports

```python
builder.save_report_canonical(report_connector_name=builder.REPORT_SCHEMA, report=builder.report_canonical_schema(stylise=False))
builder.save_report_canonical(report_connector_name=builder.REPORT_CATALOG, report=builder.report_column_catalog(stylise=False))
```
